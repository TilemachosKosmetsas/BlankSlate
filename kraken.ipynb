```python
#preparing euro deposits dataset
euro=init_dataset[init_dataset['asset']== 'ZEUR'  ]
euro_deposits=euro[euro['type'] == 'deposit']
euro_deposits=euro_deposits.dropna()

#my total amount of Eur deposits.
deuro=euro_deposits['amount'].sum()
deuro
```

```python
#preparing euro withdrawals dataset
euro_withdrawals=euro[euro['type'] == 'withdrawal']
euro_withdrawals=euro_withdrawals.dropna()

#my total amount of Eur withdrawals.
weuro=euro_withdrawals['amount'].sum()
weuro
```

```python
#Preparing xrp withdrawal dataset
xrp_init=init_dataset[init_dataset['asset']== 'XXRP'  ]
xrp_withdrawals=xrp_init[xrp_init['type'] == 'withdrawal']
#xrp_withdrawals.dropna()['amount']
xrp_wit = xrp_withdrawals.dropna()['amount']
#xrp_withdrawals.dropna()['amount'].sum()
xrp_wit 
```

```python
#!pip install yfinance  
import yfinance as yf
import numpy as np
xrp_euro_prices=yf.download('XRP-EUR',start='2017-01-01')

xrp_euro_prices['average'] = (xrp_euro_prices['High'] + xrp_euro_prices['Low'])/2

#xrp_euro average prices (based on yahoo)
xrp_euro_prices['average']
```

```python
xrp_wit.index = pd.to_datetime(xrp_wit.index)
y =xrp_wit.index.date
xrpprices=pd.DataFrame()

for i in y:
    dlol=pd.DataFrame((xrp_euro_prices['average'][xrp_euro_prices.index.date == i]))
    
    xrpprices=xrpprices.append(dlol)
    
#manually adding the 3 missing values of xrp (at which time yahoo does not provide the xrp-euro prices)
missingxrp=pd.DataFrame([0.2871,0.2871,0.193],columns=['average'],index=pd.to_datetime(['2017-06-23','2017-06-23','2017-09-03']))
missingxrp.index.rename('Date',inplace=True)
#missingxrp 

xrppricesfinal = pd.concat([ missingxrp,xrpprices])

xrp_wit2=pd.DataFrame(xrp_wit)
xrp_wit2.index= xrp_wit2.index.date

#calculating my xpr withdrawals in fiat terms (approximately: using average daily values).
xrp_wit2=pd.DataFrame(xrp_wit)
xrp_wit2.index= xrp_wit2.index.date

calc=pd.concat([xrp_wit2,xrppricesfinal],axis=1)
calc
```

```python
#euros withdrawed via xrp
wxrp=(calc['amount']*calc['average']).sum()
wxrp
```

```python
#euros deposited via xrp
xrp_deposits=xrp_init[xrp_init['type'] == 'deposit']
xrp_dep_final=xrp_deposits.dropna()
xrp_dep_final.index = pd.to_datetime(xrp_dep_final.index)
xrp_dep_final.index=xrp_dep_final.index.date
xrp_dep_amount = pd.DataFrame(xrp_dep_final['amount'])
xrp_dep_amount['average']= [0.73,0.276]

#xrp_dep_amount
dxrp=(xrp_dep_amount['amount']*xrp_dep_amount['average']).sum()
dxrp
```

```python
#Downloading eth euro prices pair and creating an average column
eth_euro_prices=yf.download('ETH-EUR',start='2017-01-01')

#my eth withdrawals
eth_init=init_dataset[init_dataset['asset']== 'XETH'  ]
eth_withdrawals=eth_init[eth_init['type'] == 'withdrawal']

eth_wit = eth_withdrawals.dropna()['amount']

eth_wit_final=pd.DataFrame(eth_wit) 
eth_wit_final.index=pd.to_datetime(eth_wit_final.index)
eth_wit_final.index=eth_wit_final.index.date
#eth_wit_final

eth_euro_prices['average']=(eth_euro_prices['High'] + eth_euro_prices['Low'])/2
#eth_euro_prices
```

```python
#debugging ( prices correctly correspond when using join date--> eth price)
#pd.DataFrame(eth_euro_prices['average'])[eth_euro_prices.index=='2018-04-26']
eth_wit_final.index=pd.to_datetime(eth_wit_final.index)
eth_wit_final.index

#pd.concat([eth_wit_final,eth_euro_prices['average']],axis=1)
calc2=eth_wit_final.join(eth_euro_prices['average'],how='inner')

calc2.loc[pd.to_datetime('2017-11-08')]=['-0.39500','266.49']
calc2.loc[pd.to_datetime('2017-08-19')]=[pd.to_numeric('-0.39500'),'263.09']
calc2=calc2.sort_index()
calc2=pd.DataFrame(calc2)
#calc2

#converting the inserted values to float (1d - array)
calc2.iloc[1,0:2]=pd.to_numeric(calc2.iloc[1,0:2])
calc2.iloc[0,0:2]=pd.to_numeric(calc2.iloc[0,0:2])

##euros withdrawed via eth
weth=(calc2['amount']*calc2['average']).sum()
weth
```

```python
#my eth deposits

eth_deposits=eth_init[eth_init['type'] == 'deposit']

eth_dep = eth_deposits.dropna()['amount']

eth_dep_final=pd.DataFrame(eth_dep) 
eth_dep_final.index=pd.to_datetime(eth_dep_final.index)
eth_dep_final.index=eth_dep_final.index.date
#eth_dep_final

calc3=eth_dep_final.join(eth_euro_prices['average'],how='inner')
calc3.loc[pd.to_datetime('2017-06-12')]=['0.194154','358.47']
calc3.loc[pd.to_datetime('2017-06-23')]=['0.193301','306.57']
calc3=calc3.sort_index()
calc3=pd.DataFrame(calc3)
calc3.iloc[1,0:2]=pd.to_numeric(calc3.iloc[1,0:2])
calc3.iloc[0,0:2]=pd.to_numeric(calc3.iloc[0,0:2])
calc3
```

```python
#my euro deposits ( in eth terms)
deth=(calc3['amount']*calc3['average']).sum()
deth
```

```python
#Downloading eos euro prices pair and creating an average column
eos_euro_prices=yf.download('EOS-EUR',start='2017-01-01')

#Calculating euro withdrawed via eos
eoswit=init_dataset[(init_dataset['asset'] == 'EOS') & (init_dataset['type'] == 'withdrawal')].dropna()
eoswit=eoswit['amount']
eoswit = pd.DataFrame(eoswit)

eoswit.index=pd.to_datetime(eoswit.index)
eoswit.index=eoswit.index.date
#eoswit

#eos eur prices
eos_euro_prices['average']=(eos_euro_prices['High'] + eos_euro_prices['Low'])/2
#eos_euro_prices['average']

#the dataset of my withdrawed eos and their corresponding fiat (EUro) values at the time (approximated with Yahoo finance).
calc4=pd.DataFrame(eoswit).join(eos_euro_prices['average'],how='inner')
calc4
```

```python
#Euro withdrawed via eos:
weos=(calc4['amount']*calc4['average']).sum()
weos
```

```python
#Calculating euro deposited via eos
eosdep=init_dataset[(init_dataset['asset'] == 'EOS') & (init_dataset['type'] == 'deposit')].dropna()
eosdep=eosdep['amount']
eosdep = pd.DataFrame(eosdep)

eosdep.index=pd.to_datetime(eosdep.index)
eosdep.index=eosdep.index.date
#eosdep

calc5=pd.DataFrame(eosdep).join(eos_euro_prices['average'],how='inner')
#calc5

deos=(calc5['amount']*calc5['average']).sum()
deos
```

```python
#Lastly finding the value of deposited dash in euro
dashdep=init_dataset[(init_dataset['asset'] == 'DASH') & (init_dataset['type'] == 'deposit')].dropna()
dashdep=dashdep['amount']
dashdep = pd.DataFrame(dashdep)

dashdep.index=pd.to_datetime(dashdep.index)
dashdep.index=dashdep.index.date
#dashdep
#my deposited value in euro is:
ddash=dashdep['amount'][0]*121.5
#ddash
```
